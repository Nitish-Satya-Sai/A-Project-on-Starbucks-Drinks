### Working for the "varies" value

## In caffeine column, few values are imputed with "varies" and "Varies". In general, there are few Starbucks drinks in which caffeine value will be varied

### It is difficult to impute caffeine values for such drinks, So I decided to drop the records.

```python
indextr = star_bucks[ (star_bucks["Caffeine (mg)"]=="varies") | (star_bucks["Caffeine (mg)"]=="Varies") ].index
```

```python
type(indextr)
```

```python
star_bucks.drop(indextr,inplace=True)
```

## Imputing the one visible Null/Missing value

```python
star_bucks[star_bucks["Caffeine (mg)"].isna()]
```

```python
mask1=star_bucks["Beverage_category"]=="Shaken Iced Beverages"
mask2=star_bucks["Beverage"]=="Iced Brewed Coffee (With Milk & Classic Syrup)"
mask3=star_bucks["Beverage_prep"]=="2% Milk"
helper_for_imputation=star_bucks.where(mask1).where(mask2).where(mask3).dropna(axis=0)
```

```python
helper_for_imputation
```

```python
star_bucks_copy = star_bucks.copy()
```

```python
star_bucks_copy["Caffeine (mg)"]=star_bucks["Caffeine (mg)"].fillna(helper_for_imputation["Caffeine (mg)"].mean())
```

```python
star_bucks_copy.isnull().sum()
```
